```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import csv

data = np.genfromtxt(r'C:\Users\CRISTIAN\Desktop\datos2.csv', delimiter=';')
print(data)
```

```python
data[30]=[-0.5,1,-1.5]
data[31]=[-0.5,0,0.5]
print(data)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(data[:,0], data[:,1], data[:,2], c='g', marker='o')
```

```python
krango=range(1,10)
sse=[] #suma errores al cuadrado
for k in krango:
  kmeans=KMeans(n_clusters=k).fit(data)
  sse.append(kmeans.inertia_)
```

```python
plt.xlabel('X')
plt.ylabel('sse')
plt.plot(krango,sse)
```

```python
data[-4:,]
centroids1=data[-4:,]
kmeans = KMeans(n_clusters=4,n_init=1,init=data[-4:,]).fit(data)
#kmeans = KMeans(n_clusters=4).fit(data)
#centroids = kmeans.cluster_centers_
#print(centroids)
#print(kmeans)
kmeans.labels_
```

```python
labels = kmeans.predict(data)
 # Getting the cluster centers
#C = kmeans.cluster_centers_
colores=['red','green','blue','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
    
from mpl_toolkits.mplot3d import Axes3D    
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=asignar,s=55)
#ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=400)
ax.scatter(centroids1[:,0],centroids1[:,1],centroids1[:,2],marker='*', c=colores, s=400)
plt.show()
```

```python
registros=np.array([data[3],data[9],data[19],data[29]])
print(registros)
```

```python
kmeans = KMeans(n_clusters=4,n_init=1,init=registros).fit(data)
kmeans.labels_
```

```python
from mpl_toolkits.mplot3d import Axes3D    
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=kmeans.labels_.astype(float),s=55)
ax.scatter(registros[:,0],registros[:,1],registros[:,2],c='red',marker='*',s=500)
```

```python
pr=np.array([[3,3,3],[1,3,3],[3,3,2],[1,2,2],[0,3,3]])
print(pr)
#kmeans.predict(pr)
kmeans.fit_predict(pr)
```
